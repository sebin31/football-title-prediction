```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
df['GD']=(df['GF'] - df['GA'])
df['GD']
```

```python
df['winner']=(df['PTS']==df['PTS'].max()).astype(int)
```

```python
df['GD']=(df['GF'] - df['GA'])
features=['M','W', 'D', 'L', 'GF','GA', 'PTS', 'xG_Team', 'xGA_Team','GD']
x=df[features]
y=df['winner']
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
rf=RandomForestClassifier()
```

```python
rf.fit(x_train, y_train)
```

```python
y_pred=rf.predict(x_test)
y_pred
```

```python
print("y_test dtype:", y_test.dtype)
print("y_pred dtype:", type(y_pred))
print("First few values of y_test:", y_test[:10])
print("First few values of y_pred:", y_pred[:10])
```

```python
print('Accuracy',accuracy_score(y_test, y_pred))
print(classification_report(y_test,y_pred))
```

```python
df = df.groupby('Team', as_index=False)['Win Probability'].max()
df = df.sort_values(by='Win Probability', ascending=False)
print(df)
print(df[['Team', 'Win Probability']].to_string(index=False))
```

```python
df['Win Probability'] = df['Win Probability'] * 100  # Convert to percentage
df['Win Probability'] = df['Win Probability'].round(2)  # Round for cleaner display
print(df[['Team', 'Win Probability']])
```

```python
df = df[df['Win Probability'] > 0]  # Keep only teams with some chance
print(df)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.barh(df['Team'], df['Win Probability'], color='skyblue')
plt.xlabel('Win Probability (%)')
plt.ylabel('Team')
plt.title('Predicted Chances of Winning the Title')
plt.gca().invert_yaxis()  # Put highest on top
plt.show()
```
